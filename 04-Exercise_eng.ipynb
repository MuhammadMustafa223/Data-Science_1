````python
# Clean megesort.py to follow clean ode principles
"""Implementation of the Merge Sort algorithm with performance visualization."""
from typing import List
import matplotlib.pyplot as plt
import time
import random

def merge_sort(arr: List[int]) -> List[int]:
    """
    Sorts a list of integers using the Merge Sort algorithm (divide and conquer).

    Args:
        arr (List[int]): List of integers to sort.

    Returns:
        List[int]: Sorted list in ascending order.

    Raises:
        TypeError: If input is not a list or contains non-comparable elements.
    """
    if not isinstance(arr, list):
        raise TypeError("Input must be a list")
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    return merge(left, right)

def merge(left: List[int], right: List[int]) -> List[int]:
    """
    Merges two sorted lists into a single sorted list.

    Args:
        left (List[int]): First sorted list.
        right (List[int]): Second sorted list.

    Returns:
        List[int]: Merged sorted list.
    """
    result = []
    left_idx, right_idx = 0, 0
    while left_idx < len(left) and right_idx < len(right):
        if left[left_idx] <= right[right_idx]:
            result.append(left[left_idx])
            left_idx += 1
        else:
            result.append(right[right_idx])
            right_idx += 1
    result.extend(left[left_idx:])
    result.extend(right[right_idx:])
    return result

def visualize_sorting_performance(size: int = 1000) -> None:
    """
    Visualizes the sorted output of a random list and measures execution time.

    Args:
        size (int): Size of the random list to sort (default: 1000).
    """
    # Generate random list
    input_list = [random.randint(0, 1000) for _ in range(size)]
    # Measure sorting time
    start_time = time.time()
    sorted_list = merge_sort(input_list)
    end_time = time.time()
    # Plot sorted list
    plt.figure(figsize=(10, 5))
    plt.plot(range(len(sorted_list)), sorted_list, label="Sorted Values")
    plt.xlabel("Index")
    plt.ylabel("Value")
    plt.title(f"Merge Sort Output (n={size})")
    plt.legend()
    plt.grid(True)
    plt.show()
    print(f"Sorting time: {end_time - start_time:.4f} seconds")

if __name__ == "__main__":
    visualize_sorting_performance()
````

b) [10%] Implement find_single_element

````python
from typing import List

def find_single_element(arr: List[int]) -> int:
    """
    Finds the element that appears exactly once in an array where all others appear twice.

    Uses XOR: paired elements cancel out, leaving the single element.
    
    Args:
        arr (List[int]): List of integers where all but one element appears twice.
    
    Returns:
        int: The element appearing exactly once.
    
    Raises:
        TypeError: If input is not a list or has even length.
    """
    if not isinstance(arr, list):
        raise TypeError("Input must be a list")
    if len(arr) == 0 or len(arr) % 2 == 0:
        raise TypeError("Array must have odd length")
    
    result = 0
    for num in arr:
        result ^= num
    return result
````
