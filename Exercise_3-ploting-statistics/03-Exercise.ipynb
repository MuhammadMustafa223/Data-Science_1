b) [10%] Define a function that creates a rotation matrix depending on a rotation angle $\theta$

````python
import numpy as np

# Function to generate a 2D rotation matrix
def get_rotation_matrix(theta):
    """
    Return a 2x2 rotation matrix for a given angle theta (in radians).
    """
    rotation_matrix = np.array([
        [np.cos(theta), -np.sin(theta)],
        [np.sin(theta),  np.cos(theta)]
    ])
    return rotation_matrix

# Example usage
theta = np.radians(40)  # Convert 40 degrees to radians
r_matrix = get_rotation_matrix(theta)

# Assuming 'data' contains your 2D points
# If needed, generate sample data for demonstration:
# mean = np.array([0, 0])
# cov = np.array([[1, 0], [0, 20]])
# np.random.seed(101)
# data = np.random.multivariate_normal(mean, cov, size=2000)

# Rotate the data
data_rotated = data @ r_matrix.T  # Apply rotation
````

c) [25%] Create a plot that looks as similar as possible to the example plot:

Hints:
- Many of the functions used here were not covered in the lecture or tutorial - Google helps!
- You can easily plot the data in a scatterplot if you transpose the data array and then pass it to the plot function using `*`: `ax.scatter(*data.T)`
- The cross that shows the mean is its own plot consisting of just one point
- The transparency in the points and histograms is helpful when you want to make things visible that overlap

````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

# Set seed for reproducibility
np.random.seed(101)

# Generate original data (from part a)
mean = np.array([0, 0])
cov = np.array([[1, 0.], [0., 20]])
data = np.random.multivariate_normal(mean, cov, 2000)

# Create rotation matrix and rotate data (from part b)
theta = np.radians(40)  # 40 degree rotation
r_matrix = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta),  np.cos(theta)]
])
data_rotated = data @ r_matrix.T

# Create figure with custom layout
plt.figure(figsize=(10, 8))
gs = GridSpec(2, 2, width_ratios=[3, 1], height_ratios=[1, 3])

# Main scatter plot
ax_scatter = plt.subplot(gs[1, 0])
ax_scatter.scatter(data[:, 0], data[:, 1], alpha=0.3, s=10, label='Original')
ax_scatter.scatter(data_rotated[:, 0], data_rotated[:, 1], alpha=0.3, s=10, label='Rotated (40°)')

# Add mean markers
mean_original = data.mean(axis=0)
mean_rotated = data_rotated.mean(axis=0)
ax_scatter.scatter(*mean_original, marker='x', color='red', s=100, linewidth=2)
ax_scatter.scatter(*mean_rotated, marker='x', color='blue', s=100, linewidth=2)

ax_scatter.set_xlabel('X')
ax_scatter.set_ylabel('Y')
ax_scatter.set_title('a) Original and rotated point cloud')
ax_scatter.legend()
ax_scatter.grid(True)
ax_scatter.set_xlim(-15, 5)
ax_scatter.set_ylim(-15, 15)

# Histogram plot
ax_hist = plt.subplot(gs[1, 1], sharey=ax_scatter)
ax_hist.hist(data[:, 0], bins=30, orientation='horizontal', alpha=0.7, color='C0', label='Original')
ax_hist.hist(data_rotated[:, 0], bins=30, orientation='horizontal', alpha=0.7, color='C1', label='Rotated')
ax_hist.set_xlabel('Count')
ax_hist.set_title('b) Histogram of the x dimension')
ax_hist.legend()

plt.tight_layout()
plt.show()
````
